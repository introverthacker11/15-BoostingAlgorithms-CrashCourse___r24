### Heart Disease Dataset Variables

- **1. age:**
  - **Type**: Integer
  - **Description**: The age of the patient.
  - **Units**: Years
  - **Note**: Age is a key risk factor for heart disease.

- **2. sex:**
  - **Type**: Categorical
  - **Description**: The sex of the patient, with possible values like male or female.

- **3. cp (chest pain type):**
  - **Type**: Categorical
  - **Description**: Type of chest pain experienced by the patient. It categorizes chest pain into four types:
    - 1: Typical angina
    - 2: Atypical angina
    - 3: Non-anginal pain
    - 4: Asymptomatic

- **4. trestbps (resting blood pressure):**
  - **Type**: Integer
  - **Description**: The patient’s blood pressure when at rest (on admission to the hospital).
  - **Units**: mm Hg
  - **Note**: Blood pressure is an important indicator of heart health.

- **5. chol (serum cholesterol):**
  - **Type**: Integer
  - **Description**: Serum cholesterol level in the patient’s blood.
  - **Units**: mg/dl
  - **Note**: High cholesterol is a major risk factor for heart disease.

- **6. fbs (fasting blood sugar):**
  - **Type**: Categorical
  - **Description**: Whether the patient’s fasting blood sugar is greater than 120 mg/dl.
  - **Note**: Indicates the risk of diabetes, which is linked to heart disease.

- **7. restecg (resting electrocardiographic results):**
  - **Type**: Categorical
  - **Description**: Results from a resting electrocardiogram, which assesses heart activity. This can be:
    - Normal
    - Showing ST-T wave abnormalities
    - Showing hypertrophy

- **8. thalach (maximum heart rate achieved):**
  - **Type**: Integer
  - **Description**: The maximum heart rate achieved during exercise testing.
  - **Note**: A higher heart rate can be a positive sign, but abnormal responses may indicate issues.

- **9. exang (exercise induced angina):**
  - **Type**: Categorical
  - **Description**: Whether the patient experienced angina (chest pain) during exercise.
  - **Note**: Exercise-induced angina is a key indicator of heart problems.

- **10. oldpeak (depression induced by exercise):**
  - **Type**: Integer
  - **Description**: ST depression induced by exercise relative to rest. It’s a measure of heart stress during exercise.

- **11. slope:**
  - **Type**: Categorical
  - **Description**: The slope of the peak exercise ST segment (indicative of heart stress).
  - **Note**: This can show how the heart is responding to stress.

- **12. ca (number of major vessels colored by fluoroscopy):**
  - **Type**: Integer
  - **Description**: The number of major blood vessels (0-3) that show up as being colored by fluoroscopy.
  - **Note**: A higher number suggests healthier vessels.

- **13. thal:**
  - **Type**: Categorical
  - **Description**: Thalassemia, a blood disorder that can affect heart function. It can be:
    - Normal
    - Fixed defect
    - Reversible defect

- **14. num (diagnosis of heart disease):**
  - **Type**: Integer
  - **Description**: The target variable. It indicates whether the patient has heart disease or not:
    - 0: No heart disease
    - 1-4: Levels of heart disease severity
