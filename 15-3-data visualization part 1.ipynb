```python
columns_ = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']
df[columns_] = df[columns_].astype('str')

df['ca'] = df['ca'].astype('int')
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 10)
plt.rcParams['figure.dpi'] = 200
```

```python
columns = df.columns[:-1]

for i, col in enumerate(columns):
    plt.subplot(4 ,4 , i+1)
    sns.histplot(x = col, data = df, hue = 'num', color = sns.color_palette("Set1"))
    plt.xlabel(col)
    plt.title("{} vs target".format(col))
    plt.tight_layout()
```

```python
color_ = {0 : 'blue', 1 : 'green', 2 : 'red', 3 : 'black', 4 : 'brown'}
sns.pairplot( data = df, hue = 'num', palette = color_)
```
